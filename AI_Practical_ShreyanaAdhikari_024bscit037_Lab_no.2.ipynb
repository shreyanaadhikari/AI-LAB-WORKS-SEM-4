Pandas Basics

```python
# 1. Create a Pandas DataFrame with columns: Name, Age, City, and Salary.
data = {
'Name': ['Alcaraz', 'Federer', 'Sinner', 'Nadal', 'Williams'],
'Age': [22, 42, 23, 40, 43],
'City': ['Madrid', 'Basel', 'Rome', 'Manacor', 'Washington D.C'],
'Salary': [500000, 650000, 700000, 100000, 60000]
}
df = pd.DataFrame(data)
print(df)
```

```python
# 2. Display the first 5 rows and last 3 rows of a DataFrame.
print(df.head(5))
print(df.tail(3))
```

```python
# 3. Check the shape, data types, and basic statistics of a DataFrame using info
print(df.shape)
print(df.info())
print(df.describe())
```

```python
# 4. Select a single column and multiple columns from a DataFrame.
print(df["Name"])
print(df[["Name", "Salary"]])
```

```python
# 5. Filter rows where Salary is greater than 40000.
print(df[df["Salary"] > 40000])
```

```python
# 6. Add a new column 'Tax' which is 10% of Salary.
df["Tax"] = df["Salary"] * 0.10
print(df)
```

```python
# 7. Sort the DataFrame by Age in ascending order.
print(df.sort_values("Age"))
```

```python
# 8. Find and count missing values in each column using isnull().sum().
print(df.isnull().sum())
```

```python
# 9. Replace missing values in a numeric column with the column mean.
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
print(df)
```

```python
# 10. Group data by City and find the average Salary per City.
print(df.groupby("City")["Salary"].mean())
```

Matplotlib Visualization

```python
# 1. Plot a simple line graph showing numbers 1 to 10 on both axes.
x = np.arange(1, 11)
y = np.arange(1, 11)
plt.title ("Line Graph on both axes")
plt.plot(x, y)
plt.show()
```

```python
# 2. Create a bar chart showing marks of 5 students.
students = ['Norris', 'Verstappen', 'Alonso', 'Hamilton', 'Vettel']
marks = [97, 98, 55, 75, 80]
plt.bar(students, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Bar Chart of Student Marks")
plt.show()
```

```python
# 3. Draw a pie chart showing percentage of students in 3 departments.
departments = ['Art History', 'Colour Theory', 'Fine Arts']
students = [700, 300, 450]
plt.pie(students, labels=departments, autopct= '%.1f%%')
plt.title("Pie Chart of Students in different Departments")
plt.show()
```

```python
# 4. Plot a histogram for 100 random numbers to see the distribution.
data = np.random.randn(100)
plt.hist(data , edgecolor='black')
plt.title("Histogram of Various Random Data")
plt.show()
```

```python
# 5. Create a scatter plot between Age and Salary from a sample dataset.
plt.scatter(df["Age"], df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Scatter Plot of Age vs Salary (Sample Data)")
plt.show()
```

```python
# 6. Plot two lines on the same graph and add a legend.
x = [5, 10, 15, 20, 25]
plt.plot(x, x, label="Line #1")
plt.plot(x, [i*2 for i in x], label="Line #2")
plt.legend()
plt.show()
```

```python
# 7. Add title, x-label, y-label, and grid to any chart.
x = [4, 2, 7, 8, 10]
y = [1, 9, 2, 3, 14]
plt.plot(x, y)
plt.title("Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
# 8. Use subplots to display a line chart and a bar chart side by side.
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Line Chart")
plt.subplot(1, 2, 2)
plt.bar(["A", "B", "C"], [20, 30, 45])
plt.title("*Bar Chart*")
plt.show()
```

Reading & Cleaning CSV Files

```python
# 1. Read a CSV file using pandas read_csv() and display its contents.
df = pd.read_csv("Copy of Automobile(1).csv")
print(df)
```

```python
# 2. Print the column names, number of rows, and number of columns.
print("Column Names:", df.columns)
print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])
```

```python
# 3. Display the first 5 rows and check for missing values.
print("First 5 rows:", df.head(5))
print("Missing values:", df.isnull().sum())
```

```python
# 4. Drop rows that contain any missing values using dropna().
df = df.dropna()
print(df)
```

```python
# 5. Fill missing numeric values with the column mean using fillna().
df = df.fillna(df.mean(numeric_only=True))
print(df)
```

```python
# 6. Remove duplicate rows from the dataset.
df = df.drop_duplicates()
print(df)
```

```python
# 7. Rename column headers to lowercase using str.lower().
df.columns = df.columns.str.lower()
print(df.columns)
```

```python
# 8. Convert a categorical column (e.g. Gender: Male/Female) into numbers using
df["number_of_doors"] = df["number_of_doors"].map({"two": 2,"four": 4})
print(df[["number_of_doors"]].head())
```

```python
# 9. Select only numeric columns and print their correlation using corr().
print(df.select_dtypes(include="number").corr())
```

```python
# 10. Save the cleaned DataFrame to a new CSV file using to_csv().
df.to_csv("new_data.csv", index=False)
print("File has been saved")
```
